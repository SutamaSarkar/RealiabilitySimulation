```python
import random
import matplotlib.pyplot as plt
import math

lamda1=0.001
lamda2=0.0012
beta=1.5
theta=1000

t1=100
t2=500
t3=1000

relcomp1=relcomp2=relcomp3=math.exp(-lamda1*t1)
relcomp4=relcomp5=relcomp6=math.exp(-(t1/theta)**beta)
relcomp7=relcomp8=math.exp(-lamda2*t1)


systemrel1= (1-((1-relcomp1)*(1-relcomp2)*(1-relcomp3)))*(1-(1-((1-((1-relcomp4)*(1-relcomp5)))*relcomp6))*(1-(relcomp7*relcomp8)))

#relcomp1=relcomp2=relcomp3=math.exp(-lamda1*t2)
#relcomp4=relcomp5=relcomp6=math.exp(-(t2/theta)**beta)
#relcomp7=relcomp8=math.exp(-lamda2*t2)
print("____________________________________________________")

print(f"R1(100)=R2(100)=R3(100) = {relcomp3:.4f}")
print(f"R4(100)=R5(100)=R6(100) = {relcomp6:.4f}")
print(f"R7(100)=R8(100) = {relcomp8:.4f}")

print(f"Analytical System Reliability(100)= {systemrel1:.4f}")

print("____________________________________________________")

relcomp12=relcomp22=relcomp32=math.exp(-lamda1*t2)
relcomp42=relcomp52=relcomp62=math.exp(-(t2/theta)**beta)
relcomp72=relcomp82=math.exp(-lamda2*t2)
systemrel2= (1-((1-relcomp12)*(1-relcomp22)*(1-relcomp32)))*(1-(1-((1-((1-relcomp42)*(1-relcomp52)))*relcomp62))*(1-(relcomp72*relcomp82)))

print(f"R1(500)=R2(500)=R3(500) = {relcomp32:.4f}")
print(f"R4(500)=R5(500)=R6(500) = {relcomp62:.4f}")
print(f"R7(500)=R8(500) = {relcomp82:.4f}")

print(f"Analytical System Reliability(500)= {systemrel2:.4f}")

print("____________________________________________________")

relcomp13=relcomp23=relcomp33=math.exp(-lamda1*t3)
relcomp43=relcomp53=relcomp63=math.exp(-(t3/theta)**beta)
relcomp73=relcomp83=math.exp(-lamda2*t3)

systemrel3= (1-((1-relcomp13)*(1-relcomp23)*(1-relcomp33)))*(1-(1-((1-((1-relcomp43)*(1-relcomp53)))*relcomp63))*(1-(relcomp73*relcomp83)))

print(f"R1(1000)=R2(1000)=R3(1000) = {relcomp33:.4f}")
print(f"R4(1000)=R5(1000)=R6(1000) = {relcomp63:.4f}")
print(f"R7(1000)=R8(1000) = {relcomp83:.4f}")
print(f"Analytical System Reliability(1000)= {systemrel3:.4f}")

print("____________________________________________________")
```

```python
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(42)
T_sys = np.zeros(N)

for j in range(N):

    #subsys1
    rand_c1=np.random.uniform(a,b)
    t1=-(1/lamda1)*np.log(1-rand_c1)

    rand_c2=np.random.uniform(a,b)
    t2=-(1/lamda1)*np.log(1-rand_c2)

    rand_c3=np.random.uniform(a,b)
    t3=-(1/lamda1)*np.log(1-rand_c3)

    T1=max(t1,t2,t3)

    #subsys 3
    rand_c4=np.random.uniform(a,b)
    t4=theta*(-np.log(1-rand_c4))**(1/beta)

    rand_c5=np.random.uniform(a,b)
    t5=theta*(-np.log(1-rand_c5))**(1/beta)

    T31=max(t4,t5)

    rand_c6=np.random.uniform(a,b)
    t6=theta*(-np.log(1-rand_c6))**(1/beta)

    T32=min(T31,t6)

    #subsys2
    rand_c7=np.random.uniform(a,b)
    t7=-(1/lamda2)*np.log(1-rand_c7)

    rand_c8=np.random.uniform(a,b)
    t8=-(1/lamda2)*np.log(1-rand_c8)

    T2=min(t7,t8)


    T3=max(T32,T2)
    T_sys[j]=min(T1,T3)
    #T=min(T1,T3)
    #print(f"Row {j+1}: T = {T_sys[j]:.4f}")

for i in range(10):   # show only first 10 rows
    print(f"Row {i+1}: T = {T_sys[i]:.4f}")


#print(f"System's TTF = {T:.4f}")
print(f"Mean system TTF = {T_sys.mean():.2f}")
print(f"Median system TTF = {np.median(T_sys):.2f}")



plt.figure()
plt.hist(T_sys, bins=50, density=True)
plt.xlabel("System Time to Failure (T)")
plt.ylabel("Probability Density")
plt.title("Histogram of System Time-to-Failure")
plt.grid(True)
plt.show()
```
