```python
import random
import matplotlib.pyplot as plt
import math

R1=0.89
R2=0.91
R3=0.97
R4=0.92
R5=0.77
R6=0.86
R7=0.96
R8=0.91


subsys1=1-((1-R1)*(1-R2)*(1-R3))
subsys2=(R7*R8)
subsys3=(1-((1-R4)*(1-R5)))*R6

system=(1-((1-subsys3)*(1-subsys2)))*subsys1

print(f"Total System's Analytical Reliability = {system:.4f}")
print(f"subSystem1's Analytical Reliability = {subsys1:.4f}")
print(f"subSystem2's Analytical Reliability = {subsys2:.4f}")
print(f"subSystem3's Analytical Reliability = {subsys3:.4f}")


```

```python
N=100000
x=0
y=1
count =0
sub1=0
sub2=0
sub3=0

for i in range(1,N+1):
    a=random.uniform(x,y)
    b=random.uniform(x,y)
    c=random.uniform(x,y)
    d=random.uniform(x,y)
    e=random.uniform(x,y)
    f=random.uniform(x,y)
    g=random.uniform(x,y)
    h=random.uniform(x,y)

    comp1=(a<=R1)
    comp2=(b<=R2)
    comp3=(c<=R3)
    comp4=(d<=R4)
    comp5=(e<=R5)
    comp6=(f<=R6)
    comp7=(g<=R7)
    comp8=(h<=R8)



    if (comp1 or comp2 or comp3) and (((comp4 or comp5) and comp6)or(comp7 and comp8)):
        count+=1
    if (comp1 or comp2 or comp3):
        sub1+=1
    if (comp7 and comp8):
        sub2+=1
    if ((comp4 or comp5) and comp6):
        sub3+=1
experiment=count/N
ss1=sub1/N
ss2=sub2/N
ss3=sub3/N
print(f"Total System's Experimental Reliability = {experiment:.4f}")
print(f"subSystem1's Experimental Reliability = {ss1:.4f}")
print(f"subSystem2's Experimental Reliability = {ss2:.4f}")
print(f"subSystem3's Experimental Reliability = {ss3:.4f}")

```

```python
trial_list=[5,10,50,100,500,1000,5000,10000,50000,100000,500000,1000000,10000000]
N_list= []
experimental_list = []
sub1_list = []
sub2_list = []
sub3_list = []

print("\nTRIALS(N)\t\tEXPERIMENTAL ANSWER\t\t SUBSYS1 \t\t SUBSYS2 \t\t \tSUBSYS3")
print("______________________________________________________________________________________________________________________________________")
for N in trial_list:
    x=0
    y=1
    count =0

    for i in range(1,N+1):
        a=random.uniform(x,y)
        b=random.uniform(x,y)
        c=random.uniform(x,y)
        d=random.uniform(x,y)
        e=random.uniform(x,y)
        f=random.uniform(x,y)
        g=random.uniform(x,y)
        h=random.uniform(x,y)

        comp1=(a<=R1)
        comp2=(b<=R2)
        comp3=(c<=R3)
        comp4=(d<=R4)
        comp5=(e<=R5)
        comp6=(f<=R6)
        comp7=(g<=R7)
        comp8=(h<=R8)



        if (comp1 or comp2 or comp3) and (((comp4 or comp5) and comp6)or(comp7 and comp8)):
            count+=1
        if (comp1 or comp2 or comp3):
            sub1+=1
        if (comp7 and comp8):
            sub2+=1
        if ((comp4 or comp5) and comp6):
            sub3+=1

    experiment=count/N
    ss1=sub1/N
    ss2=sub2/N
    ss3=sub3/N
    N_list.append(N)
    experimental_list.append(experiment)
    sub1_list.append(ss1)
    sub2_list.append(ss2)
    sub3_list.append(ss3)

for i in range(len(N_list)):
    print(f"{N_list[i]}\t\t\t\t{experimental_list[i]}\t\t\t\t{sub1_list[i]}\t\t\t{sub2_list[i]}\t\t\t\t{sub3_list[i]:.6f}")

plt.plot(N_list, experimental_list,
         color="red", marker="o",
         label="Experimental Reliability ")

plt.axhline(system,
            color="green", linestyle="--",
            label="Analytical Reliability system")
plt.plot(N_list, sub1_list,
         color="black", marker="o",
         label="Experimental Reliability ")
plt.axhline(subsys1,
            color="pink", linestyle="--",
            label="Analytical Reliability subsys1")
plt.plot(N_list, sub2_list,
         color="blue", marker="o",
         label="Experimental Reliability ")
plt.axhline(subsys2,
            color="magenta", linestyle="--",
            label="Analytical Reliability subsys2")
plt.plot(N_list, sub1_list,
         color="orange", marker="o",
         label="Experimental Reliability ")
plt.axhline(subsys3,
            color="violet", linestyle="--",
            label="Analytical Reliability subsys3")

plt.xscale("log")
plt.xlabel("Number of Trials (log scale)")
plt.ylabel("Estimated Reliability")

plt.title("Monte Carlo Convergence of series")
plt.legend()
plt.grid(True)
plt.show()
```
